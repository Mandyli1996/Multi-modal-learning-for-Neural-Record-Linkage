```python
import pickle as pkl
import pandas as pd
```

```python
with open('data_baselinecosine.map', 'rb') as f:
    data_base = pkl.load(f)
with open('data_Deeper_img_decomcosine.map', 'rb') as f:
    data_decom = pkl.load(f)
with open('data_Deeper_imgcosine.map', 'rb') as f:
    data_img = pkl.load(f)
with open('data_vbi.map', 'rb') as f:
    data_vbi = pkl.load(f)
# with open('data_vbi.map', 'rb') as f:
#     data_hybrid = pkl.load(f)
    
```

```python
def csv(data_base):
    a = [[i] for i in data_base['data'][0]]
    b =  [[i] for i in data_base['data'][6]]
    c =  [[i[0]] for i in data_base['label']]
    d = [[i[0]] for i in data_base['prediction']]
    df = pd.DataFrame(a )
    df['text_1'] = pd.DataFrame(a )
    df['price_1'] =pd.DataFrame(data_base['data'][1])
    df['lat_1'] = pd.DataFrame(data_base['data'][2])
    df['lon_1'] = pd.DataFrame(data_base['data'][3])
    df['category_1'] = pd.DataFrame(data_base['data'][4])
    df['region_1'] = pd.DataFrame(data_base['data'][5])
    df['text_2'] = pd.DataFrame(b )
    df['price_2'] =pd.DataFrame(data_base['data'][7])
    df['lat_2'] = pd.DataFrame(data_base['data'][8])
    df['lon_2'] = pd.DataFrame(data_base['data'][9])
    df['category_2'] = pd.DataFrame(data_base['data'][10])
    df['region_2'] = pd.DataFrame(data_base['data'][11])
    df['label'] = pd.DataFrame(c)
    df['prediction'] = pd.DataFrame(d)
    return df
def csv_img(data_base):
    a = [[i] for i in data_base['data'][0]]
    b =  [[i] for i in data_base['data'][7]]
    c =  [[i[0]] for i in data_base['label']]
    d = [[i[0]] for i in data_base['prediction']]
    df = pd.DataFrame(a )
    df['text_1'] = pd.DataFrame(a )
    df['price_1'] =pd.DataFrame(data_base['data'][1])
    df['lat_1'] = pd.DataFrame(data_base['data'][2])
    df['lon_1'] = pd.DataFrame(data_base['data'][3])
    df['category_1'] = pd.DataFrame(data_base['data'][4])
    df['region_1'] = pd.DataFrame(data_base['data'][5])
    df['image_1'] = pd.DataFrame(data_base['data'][6])
    df['text_2'] = pd.DataFrame(b )
    df['price_2'] =pd.DataFrame(data_base['data'][8])
    df['lat_2'] = pd.DataFrame(data_base['data'][9])
    df['lon_2'] = pd.DataFrame(data_base['data'][10])
    df['category_2'] = pd.DataFrame(data_base['data'][11])
    df['region_2'] = pd.DataFrame(data_base['data'][12])
    df['image_2'] = pd.DataFrame(data_base['data'][13])
    
    df['label'] = pd.DataFrame(c)
    df['prediction'] = pd.DataFrame(d)
    return df
```

```python
df_base = csv(data_base)[:1000]
df_img = csv_img(data_img)[:1000]
df_decom = csv_img(data_decom)[:1000]
df_vbi = csv_img(data_vbi)[:1000]
#df_hybrid = csv_img(data_hybrid)[:500]
```

```python
df_base.to_csv('prediction_baseline1.csv')
df_img.to_csv('prediction_baseline1img.csv')
df_decom.to_csv('prediction_decom.csv')
df_vbi.to_csv('prediction_vbi.csv')
```

```python
df_total = csv_img(data_img)[:1000]
df_total['pre_base'] =  df_base['prediction']
df_total['pre_decom'] =  df_decom['prediction']
#df_total['pre_hybrid'] =  df_bybrid['prediction']
df_total['pre_vbi'] =  df_vbi['prediction']
```

```python
temp = []
import numpy as np
for i in range(len(df_total)):
    if df_total['label'][i] == 0.0:
        array = [df_total['prediction'][i], df_total['pre_base'][i], df_total['pre_decom'][i], df_total['pre_vbi'][i] ]
        if min(array)<0.5:
            temp_min = array.index(min(array))
            temp.append(temp_min)
        else:
            temp.append(999)
    if df_total['label'][i] == 1.0:
        array = [df_total['prediction'][i], df_total['pre_base'][i], df_total['pre_decom'][i], df_total['pre_vbi'][i] ]
        if max(array)>0.5:
            temp_max = array.index(max(array))
            temp.append(temp_max)  
        else:
            temp.append(777)
df_total['best_model'] = pd.DataFrame(np.array(temp))
df_total.to_csv('comparision.csv')
```
